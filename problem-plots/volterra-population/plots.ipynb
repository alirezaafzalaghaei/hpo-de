```python
import random
import re
import subprocess

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scienceplots
from scipy.stats import randint, uniform
from sklearn.linear_model import HuberRegressor
from tqdm import tqdm, trange

plt.style.use("science")
```

```python
mappings = ["Logarithmic", "Algebraic", "Exponential"]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for myind, Pattern in enumerate(["1", "2", "3"]):
    index = int("5" in Pattern)
    # Read the text file
    with open("runs-random-only-ab-%s" % Pattern, "r") as f:
        lines = f.readlines()

    # Define regular expressions to extract values from each line
    pattern = r"\[([-0-9.e+, ]+)\] ([0-9.e+-]+)"

    data = []

    # Extract data from each line using regular expressions
    for line in lines:
        match = re.match(pattern, line)
        if match:
            values = [float(x) for x in match.group(1).split(", ")]
            target = float(match.group(2))
            data.append(values + [target])

    columns = ["L", "mapping", "alpha", "beta", "lmbda", "err"]
    df = pd.DataFrame(data, columns=columns)

    X = df[["alpha", "beta"]].to_numpy()
    Z = df["err"].to_numpy()

    indx = Z < 100 
    Xc = X[indx]
    Zc = Z[indx]

    Xe = X[~indx]
    Ze = Z[~indx]
    ax = axes[myind]

    sc = ax.scatter(Xc[:, 0], Xc[:, 1], c=np.log10(Zc), cmap="viridis", s=12)
    
    fig.colorbar(sc, label=r"$\log_{10}(\mathcal{E})$")
    sc = ax.scatter(Xe[:, 0], Xe[:, 1], c="red", marker="s", s=6)
    ax.set_xlabel(r"$\alpha$")
    ax.set_ylabel(r"$\beta$")
    ax.set_title("%10s Mapping" % (mappings[int(Pattern) - 1],))

fig.savefig("volterra-hyperspace-alpha-beta.pdf", bbox_inches="tight")
fig.subplots_adjust(right=0.8)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for myind, Pattern in enumerate(["1", "2", "3"]):
    index = int("5" in Pattern)
    # Read the text file
    with open("runs-random-only-ab-%s" % Pattern, "r") as f:
        lines = f.readlines()

    # Define regular expressions to extract values from each line
    pattern = r"\[([-0-9.e+, ]+)\] ([0-9.e+-]+)"

    data = []

    # Extract data from each line using regular expressions
    for line in lines:
        match = re.match(pattern, line)
        if match:
            values = [float(x) for x in match.group(1).split(", ")]
            target = float(match.group(2))
            data.append(values + [target])

    # Create a DataFrame
    columns = ["L", "mapping", "alpha", "beta", "lmbda", "err"]
    df = pd.DataFrame(data, columns=columns)
    
    X = df[["alpha", "beta"]].to_numpy()
    Z = df["err"].to_numpy()

    indx = Z <= 100
    Xc = X[indx]
    Zc = Z[indx]

    Xe = X[~indx]
    Ze = Z[~indx]
    ax = axes[myind]

    sc = ax.scatter(Xc[:, 0], Xc[:, 1], c=np.log10(Zc), cmap="viridis", s=12)

    indx = Z <= 0.005
    Xc = X[indx]

    model = HuberRegressor(epsilon=1)

    model.fit(Xc[:, [0]], Xc[:, 1])

    # Create sequence of 100 numbers from 0 to 100
    xseq = np.linspace(Xc[:, 0].min(), Xc[:, 0].max(), num=100)

    # Plot regression line
    ax.plot(xseq, model.predict(xseq.reshape(-1, 1)), color="#fca103", lw=2.5)


    fig.colorbar(sc, label=r"$\log_{10}(\mathcal{E})$")
    sc = ax.scatter(Xe[:, 0], Xe[:, 1], c="red", marker="s", s=6, zorder=-1)
    ax.set_xlabel(r"$\alpha$")
    ax.set_ylabel(r"$\beta$")
    ax.set_title("%10s Mapping" % (mappings[int(Pattern) - 1],))

fig.savefig("volterra-hyperspace-alpha-beta.pdf", bbox_inches="tight")
```
